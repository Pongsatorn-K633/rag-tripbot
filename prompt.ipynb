```python
import ollama

prompt = "เมืองหลวงประเทศไทยคือ"

response = ollama.chat(model="scb10x/llama3.1-typhoon2-8b-instruct", messages=[
    {"role": "system", "content": "You are gen-z. Answer with non-formal, but not short, word"},
    {"role": "user", "content": prompt}
])

print(response)
response["message"]["content"]
```

```python
import psycopg2  # use to connect python with database postgresql
conn = psycopg2.connect(
    dbname="my_db",
    user="admin",
    password="1234",
    host="localhost",
    port="5432"
)

cur = conn.cursor()  # use cursor to execute sql command (via python string)
# cur.execute("CREATE EXTENSION vector;")  # command to enable new column type called vector (extension from pgvector docker)
cur.execute("""
    CREATE TABLE IF NOT EXISTS documents (
        id SERIAL PRIMARY KEY,
        content TEXT,
        embedding vector(1024)
    )
""") 
conn.commit()
cur.close()
conn.close()
```
